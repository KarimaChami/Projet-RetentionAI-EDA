```python
new_df = df.drop(columns=["EmployeeCount", "StandardHours", "Over18", "EmployeeNumber"])
new_df.shape
```

```python
new_df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
new_df.head()
```

```python
new_df.to_csv('data_clean.csv',index=False)
```

```python
num_cols = new_df.select_dtypes(include=['int64','float64']).columns
cat_cols = new_df.select_dtypes(include=['object']).columns

print("Colonnes numériques :", len(num_cols))
print("Colonnes numériques :", num_cols)
print("Colonnes catégorielles :", len(cat_cols))
print("Colonnes catégorielles :", cat_cols)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(x=col, data=new_df)
    plt.title(f"Distribution de {col}")
    plt.show()
```

```python
new_df.hist(figsize=(25,20))
plt.show()
```

```python
# plt.figure(figsize=(10,6))
# sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
# plt.show()
ordinal_cols = ['Education', 'EnvironmentSatisfaction', 'JobInvolvement','JobSatisfaction', 'PerformanceRating', 'RelationshipSatisfaction','WorkLifeBalance']

num_features = [col for col in num_cols if col not in ordinal_cols]
plt.figure(figsize=(12,8))
sns.heatmap(new_df[num_features].corr(), annot=True, cmap='coolwarm')
plt.title("Corrélation entre les vraies variables numériques")
plt.show()

```
